```python
import cv2
import os

# Input and output folders
input_folder = "digits"      # folder containing PNG images
output_folder = "jpg_images"     # folder to save JPG images

# Create output folder if not exists
os.makedirs(output_folder, exist_ok=True)

# Get PNG files and sort numerically
files = [f for f in os.listdir(input_folder) if f.lower().endswith(".png")]

# Sort based on numbers in filename (assumes files like 1.png, 2.png, ...)
files = sorted(files, key=lambda x: int(os.path.splitext(x)[0]))

count = 1
for file in files:
    # Read PNG image
    img = cv2.imread(os.path.join(input_folder, file))

    # Define output JPG filename (sequential same as input order)
    output_path = os.path.join(output_folder, f"{count}.jpg")

    # Save as JPG
    cv2.imwrite(output_path, img)
    print(f"Converted {file} -> {count}.jpg")

    count += 1

print("✅ Conversion complete! JPG images are sequential as input order.")
```
